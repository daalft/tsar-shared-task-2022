```python
bert_pt = transformers.AutoTokenizer.from_pretrained("neuralmind/bert-base-portuguese-cased") # pipeline('fill-mask',model="neuralmind/bert-base-portuguese-cased", use_fast=True, top_k=10)
bert_en = transformers.AutoTokenizer.from_pretrained("bert-base-uncased") # pipeline('fill-mask',model="bert-base-uncased", use_fast=True, top_k=10)
bert_es = transformers.AutoTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-cased") # pipeline('fill-mask',model="dccuchile/bert-base-spanish-wwm-cased", use_fast=True, top_k=10)
bert_m  = transformers.AutoTokenizer.from_pretrained("bert-base-multilingual-cased") # pipeline('fill-mask',model="bert-base-multilingual-cased", use_fast=True, top_k=10)
```

```python
bert_models = {"es":bert_es,
        "pt":bert_pt,
        "en":bert_en,
        "m":bert_m}

dict_subtokens = {
    "es":{},
    "pt":{},
    "en":{},
    "m":{},
}
```

```python
for multilanguage in [False, True]:
    for lang in all_ds:
        print('Lang',lang, ("(M)" if multilanguage else ""))
        bert_tokenizer = bert_models[lang if multilanguage else "m"]
        predictions = []
        ds_file = None
        results = []
        worst_case = 0
        acum = []
        acum_unk = []
        for i, case in enumerate(all_ds[lang]):
            if case["sent"] not in dict_subtokens[lang]:
                dict_subtokens[ lang if multilanguage else "m" ][case["sent"]] = {}

            dist = []
            lst = [w for w, f in case["ans"].most_common(12)]
            for t in lst:
                input_ids = bert_tokenizer([t])['input_ids'][0][1:-1]
                vals = []
                unk = []
                for id in input_ids:
                    w = bert_tokenizer.decode([id])
                    if w.startswith("##"):
                        vals.append(1)
                    else:
                        vals.append(0)
                        if t not in dict_subtokens[lang if multilanguage else "m"][case["sent"]] and t!=w:
                            dict_subtokens[lang if multilanguage else "m"][case["sent"]][t] = w
                    unk.append(1 if w=="[UNK]" else 0)
                dist.append(sum(vals))
                acum_unk.extend(unk)
                acum.append(sum(vals)/len(vals))
            # results["S" + str(i)] = dist
            results.append(dist)
            worst_case = max(worst_case,max(dist))
        print("worst case:",worst_case)
        print("avg",np.mean(acum), np.std(acum))
        print("unk: avg",np.mean(acum_unk), "total", sum(acum_unk))
        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        bp = ax.violinplot(results)
        plt.show()

```
